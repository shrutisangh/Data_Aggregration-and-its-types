```python
import pandas as pd
# step 1  create sample dataset
data_shruti={
    'Region':['North', 'South','East','West','North', 'South','East','West'],
    'Product':['A','B','A','B','A','B','A','B'],
    'Sales':[150,800,400,600,450,80,40,80],
    'Quantity':[10,50,60,40,83,56,71,25],

}
```

```python
df_hrut=pd.DataFrame(data_shruti)
df_hrut
```

```python
# step 2 grouping and aggregration
# aggregrating sales by region
sales_by_region=df_hrut.groupby('Region')['Sales'].sum()
print(sales_by_region)
```

```python
# aggregrating sales and quantity by product(mean aggregration)
mean_by_product=df_hrut.groupby('Product')[['Sales','Quantity']].mean()
mean_by_product
```

```python
# aggrgrating count of sales by region
count_by_region=df_hrut.groupby('Region')['Sales'].count()
count_by_region
```

```python
#custom aggregration
custom_aggregration=df_hrut.groupby('Region')['Sales'].agg(['min','max'])
custom_aggregration
```

```python
# step3 mutli level aggregration
# aggregrating sales by region and product
multi_level_agg=df_hrut.groupby(['Region','Product'])['Sales'].sum()
multi_level_agg
```

```python
# step 4 reset index for multi level aggregration
multi_level_agg_reset=multi_level_agg.reset_index()
multi_level_agg_reset
```

```python
import pandas as pd

# step 1  Extend dataset with time and spatial data
data_shruti={
    'Region':['North', 'South','East','West','North', 'South','East','West'],
    'Product':['A','B','A','B','A','B','A','B'],
    'Sales':[150,800,400,600,450,80,40,80],
    'Quantity':[10,50,60,40,83,56,71,25],
    'Date':pd.to_datetime(['2024-01-01','2024-05-01','2024-06-04','2024-03-05','2024-01-09','2024-05-03','2024-02-08','2024-05-03'])
}
```

```python
df_hrut=pd.DataFrame(data_shruti)
df_hrut
```

```python
# time aggregration
#step 2 set date column as index
df_hrut.set_index('Date',inplace=True)
```

```python
# aggregrating sales by month'
minthly_sales=df_hrut.resample('M')['Sales'].sum()
minthly_sales
```

```python
# aggregrating sales by quater
quaterly_sales=df_hrut.resample('Q')['Sales'].sum()
quaterly_sales
```

```python
# aggregrating sales by year
yearly_sales=df_hrut.resample('Y')['Sales'].sum()
yearly_sales
```

```python
#reset index to restore original structure
df_hrut.reset_index(inplace=True)
```

```python
# spatial Aggegrating sales by region
sales_by_region=df_hrut.groupby('Region')['Sales'].sum()
sales_by_region
```
